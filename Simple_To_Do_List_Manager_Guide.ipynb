```python
# 1. Initialize the storage list
# Each item will be a dictionary: {'task': 'description', 'completed': False}
todo_list = []

def display_menu():
    """Displays the main menu options to the user."""
    print("\n" + "="*30)
    print("      TO-DO LIST MANAGER")
    print("="*30)
    print("1. View Tasks")
    print("2. Add Task")
    print("3. Mark Task as Complete")
    print("4. Delete Task")
    print("5. Exit")
    print("="*30)
```

```python
def view_tasks():
    """Displays all tasks in the list."""
    if not todo_list:
        print("\nüéâ Your to-do list is empty! Time to add some goals.")
        return

    print("\n--- Current To-Do List ---")
    for index, task_item in enumerate(todo_list):
        # Determine the status marker
        status = "[X]" if task_item['completed'] else "[ ]"

        # Print the task number, status, and description
        print(f"{index + 1}. {status} {task_item['task']}")
    print("-" * 25)
```

```python
def add_task():
    """Prompts the user for a new task and adds it to the list."""
    new_task = input("Enter the new task: ").strip()

    if new_task:
        todo_list.append({
            'task': new_task,
            'completed': False
        })
        print(f"‚úÖ Task '{new_task}' added.")
    else:
        print("‚ùå Task cannot be empty.")
```

```python
def mark_complete():
    """Marks a task as completed based on the user-provided number."""
    view_tasks()

    if not todo_list:
        return

    try:
        # Get the task number from the user (index is number - 1)
        task_num = int(input("Enter the number of the task to mark complete: "))
        task_index = task_num - 1

        # Input validation
        if 0 <= task_index < len(todo_list):
            if todo_list[task_index]['completed']:
                print(f"‚û°Ô∏è Task {task_num} is already complete.")
            else:
                todo_list[task_index]['completed'] = True
                print(f"ü•≥ Task {task_num} marked as complete!")
        else:
            print(f"‚ùå Invalid task number. Please enter a number between 1 and {len(todo_list)}.")

    except ValueError:
        print("‚ùå Invalid input. Please enter a valid number.")
```

```python
def delete_task():
    """Deletes a task from the list based on the user-provided number."""
    view_tasks()

    if not todo_list:
        return

    try:
        task_num = int(input("Enter the number of the task to delete: "))
        task_index = task_num - 1

        if 0 <= task_index < len(todo_list):
            deleted_task = todo_list.pop(task_index)
            print(f"üóëÔ∏è Task '{deleted_task['task']}' deleted.")
        else:
            print(f"‚ùå Invalid task number. Please enter a number between 1 and {len(todo_list)}.")

    except ValueError:
        print("‚ùå Invalid input. Please enter a valid number.")
```

```python
def main():
    """The main function to run the To-Do List Manager."""
    while True:
        display_menu()

        # Get user choice
        choice = input("Enter your choice (1-5): ").strip()

        if choice == '1':
            view_tasks()
        elif choice == '2':
            add_task()
        elif choice == '3':
            mark_complete()
        elif choice == '4':
            delete_task()
        elif choice == '5':
            print("\nüëã Exiting To-Do List Manager. Have a productive day!")
            break  # Exit the loop
        else:
            print("‚ùó Invalid choice. Please enter a number between 1 and 5.")

# Run the main function
main()
```

```python
# Import necessary modules
import random
import string
```

```python
def generate_password(length=12, use_upper=True, use_lower=True, use_digits=True, use_symbols=True):
    """
    Generates a secure, random password based on specified criteria.
    """
    # Define character sets based on input flags
    character_set = ""
    if use_upper:
        character_set += string.ascii_uppercase
    if use_lower:
        character_set += string.ascii_lowercase
    if use_digits:
        character_set += string.digits
    if use_symbols:
        # Use a common set of strong symbols
        character_set += "!@#$%^&*()-=_+"

    # Check if any character set was selected
    if not character_set:
        return "Error: Please select at least one character type."

    # Ensure the password is at least 4 characters long
    length = max(4, length)

    # Use random.sample to pick characters without replacement
    # (though for passwords, replacement is fine; sample is slightly cleaner)
    password = random.sample(character_set, length)

    # Shuffle the list to increase randomness (optional, but good practice)
    random.shuffle(password)

    # Join the list of characters into a single string
    return "".join(password)
```

```python
def check_strength(password):
    """
    Provides a simple heuristic-based strength score for a password.
    """
    score = 0
    # Length Check
    if len(password) >= 12:
        score += 3  # Strong length
    elif len(password) >= 8:
        score += 1  # Medium length

    # Character Variety Checks
    has_upper = any(c.isupper() for c in password)
    has_lower = any(c.islower() for c in password)
    has_digit = any(c.isdigit() for c in password)
    has_symbol = any(c in "!@#$%^&*()-=_+" for c in password)

    # Add points for variety
    if has_upper: score += 1
    if has_lower: score += 1
    if has_digit: score += 1
    if has_symbol: score += 1

    # Final Strength Rating
    if score >= 6:
        return "Strong üí™"
    elif score >= 4:
        return "Medium üëç"
    else:
        return "Weak ‚ö†Ô∏è"
```

```python
def main():
    """Main function to run the Password Generator and Checker."""
    print("=" * 40)
    print("PASSWORD GENERATOR AND STRENGTH CHECKER")
    print("=" * 40)

    while True:
        print("\nWhat would you like to do?")
        print("1. Generate a new password")
        print("2. Check the strength of an existing password")
        print("3. Exit")

        choice = input("Enter your choice (1-3): ").strip()

        if choice == '1':
            try:
                length = int(input("Enter desired password length (min 4): "))
                # Simple boolean inputs for criteria
                use_symbols = input("Include symbols? (y/n): ").lower() == 'y'

                # Generate and display
                new_pw = generate_password(length=length, use_symbols=use_symbols)
                print(f"\nüîë **Generated Password:** {new_pw}")

                # Check and display strength automatically
                strength = check_strength(new_pw)
                print(f"**Strength Rating:** {strength}")

            except ValueError:
                print("‚ùå Invalid input. Please enter a number for length.")

        elif choice == '2':
            pw_to_check = input("Enter the password to check: ")
            strength = check_strength(pw_to_check)
            print(f"**Strength Rating:** {strength}")

        elif choice == '3':
            print("\nüëã Exiting generator. Stay secure!")
            break

        else:
            print("‚ùó Invalid choice. Please enter 1, 2, or 3.")

# Run the main program
main()
```
